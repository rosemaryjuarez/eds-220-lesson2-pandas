# Select Rows

## selecting rows using a condition

general syntax:

```
df[conditions_on_rows]
```
example: we are interested in data after 1996


````python
df [df['year'] > 1996 ]
````

lets break it down!

- in the `df[condition_on_rows]` syntax, we have that 'condition_on_rows` = `df['year'] > 1996`


- `df['year'] > 1996` is a `pd.series` with boolean values (true false)
- pd.series wutg boolean vlaues are often called **masks**


ex: select data from years 1970 to 1979 (including 79)

````python
df [df.year.isin(range(1970,1980))]  #this is a row selection problem, so i want to select rows
````

# selecting rows using multiple conditions

we can combine multiple conditions using **or** = |, or **and**  = &

````python
#select rows with
#annual sea level rise < 0m OR annual sea level rise ? 0.8mm

df[(df['annual_sea_level_rise'] < 0) | (df['annual_sea_level_rise'] > .8) ]
````
